**Matplot**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('forestfire.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,4))
df['day'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Fire Incidents by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['temp'], bins=10, color='orange', edgecolor='black')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df.boxplot(column='wind', by='month', grid=False)
plt.title('Wind Speed by Month')
plt.suptitle('')
plt.xlabel('Month')
plt.ylabel('Wind Speed (km/h)')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['temp'], df['area'], color='red', alpha=0.6)
plt.title('Area Burned vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Area Burned (ha)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['month'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Distribution of Fires by Month')
plt.ylabel('')
plt.show()
```
